###### 2. Virtual Networking Components
***

In this presentation, we'll take a look at some of the concepts involved with network virtualization. Which is the process that allows you to connect virtual machines and devices to a network in the same way that you connect physical devices. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Virtual Machine
***

Now virtual machines, of course, do not have any physical hardware associated with them. They're just software representations of a system. 

So, they rely on the host system and its hardware for them to perform their functions. 

So virtual machines do have the ability to network with or even without using a physical network adapter. I'll clarify that in a moment, but this makes it easy to build and rebuild complex networks. And reduces the need to invest in large amounts of expensive physical hardware. 

It supports a high number of virtual networks per physical network adapter. And also supports the use of virtual Dynamic Host Configuration Protocol or DHCP. So you can still assign addresses to the virtual machines that are running on any given physical host system.

So virtual machines are going to be able to connect to the physical network through the proxy of the host network adapter. And so once it's done that, it will have the potential ability to reach all the systems that are on that network. It will have a virtual mac address, it will be able to connect through what we'll see as a virtual switch. And then reach physical switches that are out there on the network, and communicate normally. 

It's also possible that you may set up a virtual network that is internal only, that has no connection to a physical network adapter. Now, if that's the case, what are you supporting? Only that virtual machines will be able to connect to a virtual network that connects to other virtual machines. This allows you to create a safe, secure sand box, or testing environment, that will not have any conflicts with the rest of your physical network. But still allows for the interaction of network services that may be necessary to test. So, in order for us to have the functionality we want, there are several virtual components that we're going to have to put together. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Virtual Components
***

The first virtual component is a virtual switch. Now the virtual switch is run by the hypervisor of your virtualization platform. And what it does is function as a middleman between the virtual machines that are running on one end. And the physical network adapter, if we associate one with it, that then up links to the physical switch. 

So it's purely software, and it represents a network switch, so therefore, it's all about forwarding actual frames based upon Mac addresses, right? If we're talking about a switch, we know we're not thinking about IP addresses, it's at layer two, the data link layer. 

So with this, the devices that are connected into the virtual switch will have the potential to connect to each other. We can use VLANs just like we do on a physical switch to create zones of isolation based upon VLAN numbers. And we can do layer two management. So do we have certain MAC addresses that we trust? Do we prioritize certain systems, or have quality of service or thresholds in terms of bandwidth that are able to be used by certain systems? All these are switch capabilities. And then coming off of that virtual switch, we can have one or more than one physical adapter that up links to physical switches. In order to take the virtual traffic and bring it out into the physical network. Now of course, in order for this to function, each of the virtual machines must have a way to plug into that virtual switch. And they're going to use the virtual network interface controller, or the NIC, the VNIC. And so, again, a virtual machine can have one or more virtual network cards that allow them to plug into one or more virtual switches. Which then gets them connected into various LANs. It might connect them to storage area network as well. And the operating system will see the virtual NIC, not the actual physical NIC. Generally, there is going to be a specific driver and vendor that's going to be used by the hypervisor provider. Unless you're using a special type of virtualization that allows the physical network card properties to be passed through to the virtual machine. That's called SR-IOV, and it's a great technology that does give you a lot of great performance and functionality. So you've got virtual machines with virtual NIC, connecting to virtual switches, which uplink through physical network cards to physical switches. 

But there's a couple of other virtualization components to be aware of as well. You may also see the virtual router. So, routers function of course at layer three, so you can have a virtual software appliance. Which has IP addresses that is connected with its own virtual network interfaces to virtual switches. Which connect out to physical interfaces, and it can perform routing. So this means that you don't necessarily have to have a physical hardware router to provide that functionality. Which can be great, once again, for extensibility, and allowing an organization. Or part of an organization, to have its own routing function built into the system without having to dedicate a whole physical platform to it. Many of our physical routers can in fact be divided up, essentially into multiple virtual routers, to allow for a multi-tenancy experience. In a similar way, we can have virtual firewalls, which are going to, again, allow for multiple instances, like virtual machines. To be created that provide the functionality of a firewall. They block traffic that shouldn't go through, inspecting the packets, checking them to make sure that they meet security policies. And again, this can be something that is purely software running on a hypervisor. Or it may be appliance which is essentially divided up or segmented partitioned in to multiple virtual firewalls. That control policies for different departments or different customers depending on what's needed. So these are all aspects of the virtual platform that work together to help support an extensible environment, one that can grow over time.

###### 3. Hypervisors
***

Now, in this presentation we'll take a look at a core component of virtualization known as the hypervisor. 

Now, the whole idea behind a hypervisor is that we have the hardware at the lowest level, and then we have virtual machines. Well, the virtual machines need to believe that they can directly access the hardware. [A table diagram displays. The fourth base layer is the Hardware. The third layer is the Hypervisor. We need something to control access to the hardware. And this really is no different than a standard physical computer if you think about it. The operating system is what controls access to the hardware when you are running multiple applications at the same time. You can't have things that are competing for resources, if you will. There needs to be something to supervise all of this activity to say it's your turn, you have some processor time. 

Now, you have some processor time, and you have some memory, and you have some disk access, and so forth. So this is the same kind of relationship. But when it comes to the hypervisor itself, there are two types simply named type one and type two, and we'll take a look at both.

###### 4. Network Attached Storage (NAS)
***

In this presentation, we'll introduce the concept of network storage, beginning with a method known as NAS, or Network Attached Storage. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Direct Attached Storage (DAS)
***

Now, to give you a better idea of what this is all about, let me just quickly back up to its predecessor, which was DAS, Direct Attached Storage. And that is what we all tend to think about as standard storage. If you have ever gone and purchased a computer, just even a home desktop computer. And if you pull the cover off, you will find inside that computer a hard drive directly attached to the motherboard using a cable. So that is the essence of Direct Attached Storage. 

That hard drive is physically attached to that computer. And as such, that computer is the only computer that accesses that storage. Now that's all well and good, but when it comes to network environments, if that's all you have then any storage that is left over, if you will, we'll call it free storage. Effectively goes unused, it cannot be shared amongst other systems. 

Now, you can create a folder on a hard drive, and you can share it so that other systems can access that folder. But the space itself, imagine you partitioned that hard drive into two halves of equal size, 50/50. And you'd simply don't configure anything on the second partition for whatever reason. Well, that partition effectively goes to waste. It is not able to be accessed by any other server. 

Now, there wouldn't be too many instances where you would do that, but the point remains that any unused space in Direct Attached Storage remains unused. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### NAS - Removes the Direct Connection
***

So Network Attached Storage effectively removes the direct connection, and just puts all the storage together into a single unit. 

And then that unit is accessed by the clients over the network. So instead of having a directly attached hard drive, which they can still have, they can also access the Network Attached Storage. And this facilitates a very centralized method of accessing the storage. And you access it over a traditional local area network. 

Now it's referred to as file level data storage, because you effectively access whatever it is you need to access on a file per file basis. Which translates to the fact that there does need to be some kind of file system in use on the Network Attached Storage device itself. Now that's not a bad thing per se, but in terms of performance, it wasn't the best solution when you compare it to some newer approaches that we will talk about in a moment. 

But for the time being, it simply meant that there had to be a file system on the Network Attached Storage device, so that it could find the files. Very much in the same way that your own computer uses a file system to locate its own files. But again, you attached this storage mechanism directly to a standard TCP/IP network, so all clients could access it and all of the storage could be shared. So this provides data access and file storage to every system on the network. Now in the graphic, you see here [The presenter refers to a diagram that illustrates the Client Computers attached to the Local Area Network (LAN), which is in turn attached to the Network Attached Storage (NAS).] systems that look pretty much like client computers, and that's fine. 

They absolutely could, but for the most part you would probably see servers accessing the storage, but clients could. It's just a little more common that it would be the servers. So what you can accomplish with this is to possibly reduce the number of overall servers. You didn't necessarily have to dedicate servers to just storage using their own internal storage. 

So let's say I had five servers. They could all be set up as file servers, but each one only having access to its own directly attached storage and it could make that available to clients. But again, on a file per file basis, if there was any unused space, it still went unused. So let's get rid of those five file servers and put all the storage into the Network Attached Storage device. 

Now we can just consolidate everything. So it's centralized storage and makes it a lot easier in terms of accessing the data from any client system. So when it comes to accessing the storage, again, it is performed using various file level protocols. So again, this comes back to that file level access. 

And the protocols include NFS, which is the Network File System. CIFS, which is the Common Internet File System. And SMB, the Server Message Block. That one by the way, is the one that's used in Windows environments. So, that's all still well and good. You needed a way to access the files, but these protocols aren't as fast as some of the newer ones that again, we will talk about in the next presentation. But you still had to have a file system running on the Network Attached Storage device so that you could access those files. Because you requested some kind of file the NAS device needed to be able to say, okay, that file resides here, let me go retrieve it for you. So it's a little bit of a slower process, but certainly still beneficial in terms of centralizing everything. 

So you get this improved server and network performance, because again, now you just have this device that is dedicated to storage. All of the clients can access it over a traditional network. And as long as you had a solid, fast network, then accessing the files themselves was pretty good. You get the shared disk space, again, any left over free space of Direct Attached Storage could never be shared. Now all of it can be shared. It's inexpensive and easy to implement. You use your traditional networking equipment. You don't need really any specialized hardware for any of this, and it offers this increased flexibility and centralized management.

<a id="___"></a>
***
###### Conclusion - NAS step up to DAS
***

So again, this is just really the next step up from Direct Attached Storage. Now we have everything centralized, it's fairly easy to implement, it's fairly inexpensive, and just facilitates this easier management of all of the available storage within your environment.

###### 5. Storage Area Network (SAN)
***

Now, in this presentation, we'll take a look at the next level of network storage, the Storage Area Network, or SAN.

Now this is similar to network attached storage in that the servers do access the storage in a detached fashion over a network. But the similarities pretty much stop there. 

SANs are typically far more robust than network attached storage. So it's ideal for large organizations that need to store large amounts of data. And you can see that you can create the sub network of storage devices connected to the servers in a computer network including different types of storage, disk arrays, optical storage, and tape catridges. [The presenter discusses a diagram that is displayed. It illustrates clients connected to servers via a WAN and LAN. The servers then connect to SAN, which contains two switches, and SAN interacts further with various devices. These include items, such as a Disk Array, Optical Storage, and Tape Cartridge.]

But, as mentioned, it's usually far more robust. So not only do we see different types, but usually the capacity is much higher, the performance is better. And you often see these redundant connections, as you can see in the graphic above. [There are some redundant connections between the Servers; SAN; and Disk Array, Optical Storage, and Tape Cartridges.] 

Those switches in between are specific types of switches that we'll talk about in greater detail in a moment. But you often see this mesh configuration for redundancy purposes, so that if one of them goes down, we can still access the storage through a different route. [He refers to the connections facilitated by SAN.] 

And again, the other primary difference is that the storage itself is accessed. I might even go so far as to say only by the servers. 

In network attached storage, we saw that clients could directly access the storage. You virtually would never see that in a SAN configuration. The servers access the storage, the clients then get their data from the servers. 

So again, it's a much more robust configuration plus the network on the right of the servers is a different type of network. [He refers to the SAN network.] 

In network attach storage, everything is using standard Ethernet and TCP/IP but in this configuration we have different hardware we have different protocols in use to access the storage. 

Now on that note, the network on the right-hand side in the graphic above uses what's known as fiber channel technology. Now this refers to the physical media itself, fiber optic cabling, but also the protocols that are used to access the storage. They aren't standard LAN protocols. You don't use TCP/IP between the servers and the storage. So it's a combination of a few different protocols depending on the configuration. Plus, IP, Internet Protocol, if you want to connect geographically dispersed SANs together. But again, this is why you never see clients directly accessing the storage because they just don't support the protocols that are used to access the storage. It is pretty much the domain of the servers. 

So as mentioned, the protocols enable the organizations to easily share the data, and the applications with any remotely located offices, if desired, again that's not required but you can see that there is a WAN connection option there. That can allow your remote offices to access the SAN.

But with respect to the servers, accessing the storage, the protocols include iSCSI. 

Small Computer System Interface (SCSI) was the original protocol of many servers when they had direct attached storage. This is SCSI commands over an inter-network, fiber channel over IP, [FCIP, for short.] and fiber channel over Ethernet [FCoE, for short.].

So, [FCoE, for short.] again, these aren't standard networking protocols. These are protocols dedicated to accessing the storage itself. So the server access the storage, the clients access the servers and as far as the clients are concerned this SAN doesn't even really exist. They just contact the server and say, please give me my data. Server goes about its businesses and here you go. But the clients doesn't ever touch the SAN side. 

Now, another implementation that is different from network attached storage is what is known as the Logical Unit Number (LUN). 

When you have something like a disk array, there can be any number of disks in that array. And really there is no limit, you know you can have dozens, you can have hundreds. So with that type of configuration, anyone server needs to be able to say well which portion of that is my storage? What do I see as my disks? That’s where the LUNs come in. 

To just use a fairly simple example, if that disk array had let's just say 10 disks and each one was one terabyte then I could create a Logical Unit Number (LUN) that says for server one you will get two disks. As far as you are concerned that will appear to you as a single disk of two terabytes, server two you get three disks and you will see a single disk of three terabytes. Server three maybe one and half because you can absolutely use different percentages. So it's just a matter of how much of the total overall storage you want to allocate to any given server. And they see it as directly attached storage as far as the operating system is concerned. 

Now they just see a d drive or an e drive or whatever. That equates to the total amount of disk space that has been allocated to that Logical Unit Number (LUN). [Each logical disk on a SAN has a unique Logical Unit Number (LUN) for easy identification.] 

Now Logical Unit Number (LUN) masking is a technique to authorize access to that space, it's used to control which host can access which Logical Unit Number (LUN), so that server one does not try to access the Logical Unit Number (LUN) of server two, and vice versa. 

So again, there are a lot of differences with respect to a SAN when compared to network attached storage and coming back to some of the hardware, as mentioned that cabling is fiber optic cabling so the switches are whats known as fiber channel switches. So they are dedicated specific types of switches. They aren't standard Ethernet switches. 

And the servers then also have to have specific adaptors in them specifically for accessing the storage, and it's known as a Host Bus Adaptor (HBA). So you not only need a standard network interface for the severs to communicate with the rest of the LAN. You also have to have this Host Bus Adaptor (HBA). 

And you need these dedicated fiber channel switches. And all of that associated hardware. So the downside of the SAN is it's cost. It's implementation can be quite high. 

That said, its performance is very high. Accessing the storage in a SAN is done on what's known as a block basis, as opposed to a file basis.

So you don't need a file system on the SAN, so to speak. And that was something that we did see in network-attached storage. You simply contact the server and say I want this file. The server uses the appropriate storage protocols to just transfer data block by block until everything has been received and that's, much more efficient and much faster than implementing a file system on the SAN itself. 

So again, tremendous performance. Plus, you get the inherent performance of fiber connections as well. So there's no doubt that performance is there, but it is a costly solution, and represents a barrier to entry to a lot of small to medium sized organizations, which is why you do still see network attach storage. In some of the smaller environments, but SANs typically tend to be the domain of the larger enterprises.

###### 6. Connection Type
***

In our previous presentation, we introduced the concept of a Storage Area Network (SAN) and discussed how it operated over Fiber Channel.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Fiber Channel
***

So in this presentation, we'll take a closer look at Fiber Channel, because it's got a lot going for it. But there are some drawbacks, so we'll also discuss some mitigation options for some of those drawbacks.

So, with regard to the protocol itself, Fiber Channel is a technology used for high-speed transfer, allowing rates of up to 128 gigabits per second. 

And, it was designed to replace the Small Computer System Interface, or SCSI transmission interface that was used, of course, in direct attached storage. So, it's commonly used for connecting servers to storage area networks, interconnecting storage controllers and drives. And implementing connectivity between and inside a datacenter. 

But its key characteristic, it is a storage access protocol, not a LAN networking protocol, like TCP/IP. 

Now, it was also very flexible. It could cover distances of up to ten kilometers apart when using optical fiber as the physical medium. But of course it could be used for shorter distances within your office environment. So as mentioned, a lot going for it. 

But on the down side, it was expensive. The protocol itself was expensive but everything necessary to implement a fiber channel SAN, you had to have additional and specific hardware. That specific hardware often required a certain amount of expertise. So the installation and maintenance was typically expensive as well. And because you had all this additional hardware, you also had increased power and increased cooling demands. So the overall implementation of a full scale SAN tends to remain the domain of larger environments. It's still a barrier to entry of a smaller to medium size organizations. So this introduced the need to be able to implements SANs in a much simpler configuration and quite literally less expensive. 

And one way of doing so is by using Fiber Channel over Ethernet or FCoE. This still uses the storage protocol of fiber channel but allows it over standard Ethernet connections. Which, of course, almost every environment already had. 

Now with respect to the hardware, the switches that you see in graphic and the adapters, that would be in the servers, it still had to be FCoE specific. But you no longer needed dedicated equipment because the FCoE switches could handle both Fiber Channel and Ethernet. The adapters could handle both the IP traffic and Fiber Channel traffic. So I didn't need two dedicated adapters and two dedicated switches, these were all in one types of devices. And then all the cabling could remain the same as well, because it's standard Ethernet communications. So I didn't have to rip out all my cabling to implement FCoE. So again, it enables the Fiber Channel traffic to transfer across the existing Ethernet infrastructure and combines with storage and the IP protocols. So it allows for a single cable transport and a single interface. This consolidates the overall I/O, it reduces switch complexity, it reduces the infrastructure requirements with respect to again, the cables and the interface cards. And the reduced cost, the operational expenses, the power, the cooling requirements and overall management, the overall overhead associated with maintaining everything. So, you get your simplified, scaled back and less expensive SAN. 

Now another implementation which, if you will, went even a step further, was iSCSI. This still encapsulates SCSI commands but now into IP packets, which might seem the same as Fiber Channel over Ethernet. But what you don't see now, at all, is the Fiber Channel Protocol. This manages connections between devices in a network using IP, which of course is present in almost every environment. So, by not doing anything to do with fiber channel, now I don't even need the FCoE switches or the FCoE adapters. I can use my regular switches and my regular adapters. All the infrastructure that is currently in place can now be used to still access and implement SAN storage. So the components of an iSCSI SAN involve the target and the initiator. The target is, for all intents and purposes, the storage itself. It's the storage device is configured to use iSCSI. The initiators are the clients computers that access the data from the target. Now, client is a slightly misleading term here, it does not refer to a client computer like a desktop. Servers are typically the systems that access the storage. But since the server's requesting the storage from the target, then the server is a client. It's requesting something, so now you can have both software and hardware initiators. Many operating systems include the initiator as software. But you can have dedicated hardware initiators as well. But it simply initiates the I/O command sequence to say, please give me some storage. So this is required to facilitate data transmission to the storage device. But again the key component is a lack of fiber channel. Everything is done over IP so again, none of your existing infrastructure has to be taken out. So both of these represent options for implementing SAN but not a full scale SAN. But you could certainly get some benefits. They tend to perform quite well. Not as high as a full-scale SAN but at a greatly reduced cost and much less overhead. Now going the other direction, the last of these storage connection types we're going to look at is Infiniband, IB. This is another storage transport specification that's based on connecting nodes to an Infiniband fabric. Made of serial point-to-point connections between each node and the intermediate Infiniband switches. It's known for being extremely fast, though short distance delivery for I/O. And it's actually used in supercomputers. Nodes in Infiniband actually get addressed using IPv6 addresses to reference them as I/O endpoints. [Infiniband (IB) runs over copper or fiberoptic. It has Quad small form factor pluggable (QSFP) connectors, high throughput of 100+ gigabytes per second, built-in QoS and failover and is described across OSI layers 1 through 4.] Gang, that's a quick overview of a lot of these great storage transports. Not talking about how data gets written to a disk, but how we remotely connect over a SAN to make those disk requests.

***
###### 7. Jumbo Frame
***

Is quite simply a large Ethernet frame.

The standard size is only 1500 bytes. A jumbo frame is approximately 9000 bytes. 

So, it's quite simply a larger frame. Now, that might seem that it's a little bit out of place with respect to what we have been talking about so far in this course. So why are we talking about large frame sizes? Well, quite simply, when it comes to accessing storage, which we have just been discussing, you are primarily concerned with the amount of data that can exist within the frame. And this allows you to fine tune the performance of the devices that are communicating with storage to essentially support much more data. Because when we are accessing storage, that really is all we're concerned about, the data itself. There's always a fair amount of overhead involved with an Ethernet frame in terms of the source address and the destination address and the protocol information and things along those lines. So, if I can process fewer frames then I can concentrate more on the data itself.

Now, to implement jumbo frames the networking hardware has to support it. So everything must be able to accept this larger value. So quite simply, the devices need to agree on the maximum jumbo frame size. 

Now in terms of the payload, as the frames are being processed across your switches and routers, the resources can only process one frame at a time. So the CPU in those devices still is limited to saying, all right, there's a frame. Let's deal with that one. There's another one, let's deal with that one. So larger payloads can be introduced into each frame resulting in fewer frames to process. So as mentioned this increases the overall efficiency and can even help to reduce the amount of heat that is generated by the device. 

But like I said, we just want to be able to focus more on the data than on the header information and all of the overhead associated with the frame. So quite simply, fewer frames, less processing and more data. 

Now in terms of the requirements, for maximum gains, all links in the network path must be configured to use jumbo frames, and at the same Maximum Transfer Unit (MTU). So all servers, all endpoints, all have to agree and if they don't, the performance can actually decrease. Incompatible devices might drop or fragment the frames and fragmentation can result in higher CPU processing. 

So in essence, the whole process is backfiring if you do have something that is incompatible. So again, the idea behind jumbo frames typically is to implement greater efficiency with respect to the number of frames and simply containing more data. This way we have fewer frames to process and when accessing storage that's what we want. We really want to focus on the data not so much the overhead. So in a lot of storage configurations such as iSCSI where you are using standard Ethernet packets to access the storage, an increased Maximum Transfer Unit (MTU) can certainly help to improve the overall performance and the efficiency of your devices.
